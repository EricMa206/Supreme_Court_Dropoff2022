```python
import pandas as pd
import re
pd.options.display.max_rows = 300
```

```python
pokemon = pd.read_csv('Pokemon_Party.txt', delimiter = '|')
pokemon['IVs'] = pokemon['IVs'].map(lambda x: float(x.strip(' IV%')))
pokemon['Level'] = pokemon['Level'].map(lambda x: int(x.strip('Level ')))
```
